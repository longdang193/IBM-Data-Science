<h2>Functions for Plotting</h2>


````python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    """
    Plots the kernel density estimates (KDE) of two datasets on the same figure.

    Parameters:
    - RedFunction (array-like): Data for the red KDE plot.
    - BlueFunction (array-like): Data for the blue KDE plot.
    - RedName (str): Label for the red dataset.
    - BlueName (str): Label for the blue dataset.
    - Title (str): Title of the plot.
    """
    
    # Define the size of the plot in inches (width x height)
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    # Plot the KDE for the first dataset (RedFunction) with red color and label
    ax1 = sns.kdeplot(
        RedFunction,             # Data to plot
        color="r",               # Color of the KDE line (red)
        label=RedName             # Label for the legend
    )
    
    # Plot the KDE for the second dataset (BlueFunction) with blue color and label
    # The 'ax=ax1' parameter ensures it is plotted on the same axes as the first KDE
    ax2 = sns.kdeplot(
        BlueFunction,            # Data to plot
        color="b",               # Color of the KDE line (blue)
        label=BlueName,          # Label for the legend
        ax=ax1                   # Specify the axes to plot on
    )
    
    # Set the title of the plot
    plt.title(Title)
    
    # Label for the x-axis
    plt.xlabel('Price (in dollars)')
    
    # Label for the y-axis
    plt.ylabel('Proportion of Cars')
    
    # Display the legend to differentiate between the two KDE plots
    plt.legend()
    
    # Render the plot to the screen
    plt.show()
    
    # Close the figure to free up memory
    plt.close()
````

````python
def PollyPlot(xtrain, xtest, y_train, y_test, lr, poly_transform):
    """
    Plots training and testing data along with the predicted polynomial regression function.

    Parameters:
    - xtrain (array-like): Feature values for the training dataset.
    - xtest (array-like): Feature values for the testing dataset.
    - y_train (array-like): Target values for the training dataset.
    - y_test (array-like): Target values for the testing dataset.
    - lr (object): Trained linear regression model.
    - poly_transform (object): Polynomial transformation object (e.g., sklearn's PolynomialFeatures).
    """
    
    # Define the size of the plot in inches (width x height)
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    # Determine the maximum and minimum values across both training and testing feature sets
    xmax = max([xtrain.values.max(), xtest.values.max()])
    xmin = min([xtrain.values.min(), xtest.values.min()])
    
    # Create a range of x values from xmin to xmax with a step of 0.1 for plotting the predicted function
    x = np.arange(xmin, xmax, 0.1)
    
    # Plot the training data points
    plt.plot(
        xtrain,                   # Feature values for training data (x-axis)
        y_train,                  # Target values for training data (y-axis)
        'ro',                     # Red circles for training data points
        label='Training Data'     # Label for the legend
    )
    
    # Plot the testing data points
    plt.plot(
        xtest,                    # Feature values for testing data (x-axis)
        y_test,                   # Target values for testing data (y-axis)
        'go',                     # Green circles for testing data points
        label='Test Data'         # Label for the legend
    )
    
    # Plot the predicted polynomial regression function
    plt.plot(
        x,                                      # Range of x values (generated above)
        lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))),  # Predicted y values
        label='Predicted Function'             # Label for the legend
    )
    
    # Set the limits for the y-axis to ensure all data points and the prediction line are visible
    plt.ylim([-10000, 60000])
    
    # Label for the y-axis
    plt.ylabel('Price')
    
    # Label for the x-axis
    plt.xlabel('Feature')  # You can replace 'Feature' with a more specific label if applicable
    
    # Add a legend to differentiate between training data, testing data, and the predicted function
    plt.legend()
    
    # Display the plot
    plt.show()
    
    # Close the plot to free up memory (useful when generating multiple plots)
    plt.close()
````

Let's go over some examples. It turns out these differences are more apparent in Multiple Linear Regression and Polynomial Regression so we will explore overfitting in that context.</p>


Let's create <b>Multiple Linear Regression objects</b> and train the model using <b>'horsepower'</b>, <b>'curb-weight'</b>, <b>'engine-size'</b> and <b>'highway-mpg'</b> as features.

<h4>Overfitting</h4>
<p>Overfitting occurs when the model fits the noise, but not the underlying process. Therefore, when testing your model using the test set, your model does not perform as well since it is modelling noise, not the underlying process that generated the relationship. Let's create a degree 5 polynomial model.</p>


Let's use 55 percent of the data for training and the rest for testing:


````python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.45, random_state=0)
````

We will perform a <b>degree 5 polynomial transformation</b> on the feature <b>'horsepower'</b>.


````python
pr = PolynomialFeatures(degree=5)
x_train_pr = pr.fit_transform(x_train[['horsepower']])
x_test_pr = pr.fit_transform(x_test[['horsepower']])
pr
````

Now, let's create a Linear Regression model "poly" and train it.


````python
poly = LinearRegression()
poly.fit(x_train_pr, y_train)
````

We can see the output of our model using the method "predict." We assign the values to "yhat".


````python
yhat = poly.predict(x_test_pr)
yhat[0:5]
````

Let's take the first five predicted values and compare it to the actual targets.


````python
print("Predicted values:", yhat[0:4])
print("True values:", y_test[0:4].values)
````

We will use the function "PollyPlot" that we defined at the beginning of the lab to display the training data, testing data, and the predicted function.


````python
PollyPlot(x_train['horsepower'], x_test['horsepower'], y_train, y_test, poly,pr)
````

Figure 3: A polynomial regression model where red dots represent training data, green dots represent test data, and the blue line represents the model prediction.


We see that the estimated function appears to track the data but around 200 horsepower, the function begins to <b>diverge from the data points</b>.


R^2 of the training data:


````python
poly.score(x_train_pr, y_train)
````

R^2 of the test data:


````python
poly.score(x_test_pr, y_test)
````

We see the R^2 for the training data is 0.5567 while the R^2 on the test data was -29.87.  The lower the R^2, the worse the model.


Let's see how the R^2 changes on the test data for **different order polynomials** and then plot the results:


````python
Rsqu_test = []

order = [1, 2, 3, 4]
for n in order:
    pr = PolynomialFeatures(degree=n)
    
    x_train_pr = pr.fit_transform(x_train[['horsepower']])
    
    x_test_pr = pr.fit_transform(x_test[['horsepower']])    
    
    lr.fit(x_train_pr, y_train)
    
    Rsqu_test.append(lr.score(x_test_pr, y_test))

plt.plot(order, Rsqu_test)
plt.xlabel('order')
plt.ylabel('R^2')
plt.title('R^2 Using Test Data')
plt.text(3, 0.75, 'Maximum R^2 ')    
````

We see the R^2 gradually increases until an order three polynomial is used. Then, the R^2 dramatically decreases at an order four polynomial.


The following function will be used in the next section. Please run the cell below.


````python
def f(order, test_data):
    x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=test_data, random_state=0)
    pr = PolynomialFeatures(degree=order)
    x_train_pr = pr.fit_transform(x_train[['horsepower']])
    x_test_pr = pr.fit_transform(x_test[['horsepower']])
    poly = LinearRegression()
    poly.fit(x_train_pr,y_train)
    PollyPlot(x_train['horsepower'], x_test['horsepower'], y_train, y_test, poly,pr)
````

<h2 id="ref3">Part 3: Ridge Regression</h2> 


In this section, we will review Ridge Regression and see how the parameter alpha changes the model. Just a note, here our **<u>test data</u> will be used as <u>validation data</u>**.


Let's perform a degree two polynomial transformation on our data.


````python
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg','normalized-losses','symboling']])
x_test_pr=pr.fit_transform(x_test[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg','normalized-losses','symboling']])
````

Let's import  <b>Ridge</b>  from the module <b>linear models</b>.


````python
from sklearn.linear_model import Ridge
````

Let's create a Ridge regression object, setting the **regularization parameter (alpha) to 0.1**


````python
RigeModel=Ridge(alpha=1)
````

Like regular regression, you can fit the model using the method <b>fit</b>.


````python
RigeModel.fit(x_train_pr, y_train)
````

Similarly, you can obtain a prediction:


````python
yhat = RigeModel.predict(x_test_pr)
````

````python
print('predicted:', yhat[0:4])
print('test set :', y_test[0:4].values)
````

We select the value of alpha that minimizes the test error. To do so, we can use a for loop. We have also created a progress bar to see how many iterations we have completed so far.


````python
# Import necessary libraries
from tqdm import tqdm          # tqdm is used to create progress bars for loops

# Initialize lists to store R-squared scores for test and train datasets
Rsqu_test = []                 # List to store R-squared scores on the test set
Rsqu_train = []                # List to store R-squared scores on the training set
dummy1 = []                    # An unused list (could be removed or used for future purposes)

# Create an array of alpha values to test Ridge regression with
Alpha = 10 * np.array(range(0, 1000))  # Generates an array: [0, 10, 20, ..., 9990]

# Initialize a tqdm progress bar with the Alpha array
pbar = tqdm(Alpha, desc="Ridge Regression Progress", unit="alpha")

# Loop over each alpha value in the Alpha array using the progress bar
for alpha in pbar:
    # Initialize the Ridge regression model with the current alpha value
    RidgeModel = Ridge(alpha=alpha) 
    
    # Fit the Ridge regression model on the transformed training data
    RidgeModel.fit(x_train_pr, y_train)
    
    # Calculate R-squared score on the test set
    test_score = RidgeModel.score(x_test_pr, y_test)
    
    # Calculate R-squared score on the training set
    train_score = RidgeModel.score(x_train_pr, y_train)
    
    # Update the progress bar with the current test and train scores
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})
    
    # Append the current R-squared scores to their respective lists
    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
````

We can plot out the value of R^2 for different alphas:


````python
width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
````

**Figure 4**: The blue line represents the R^2 of the validation data, and the red line represents the R^2 of the training data. The x-axis represents the different values of Alpha.


Here the model is built and tested on the same data, so the training and test data are the same.

The red line in Figure 4 represents the R^2 of the training data. As alpha increases the R^2 decreases. Therefore, **as alpha increases, the model performs worse on the training data**.

The blue line represents the R^2 on the validation data. **As the value for alpha increases, the R^2 increases and converges at a point**.

````python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Assuming x_data and y_data are already defined as pandas DataFrame and Series
# Example:
# x_data = pd.DataFrame({
#     'horsepower': [130, 165, 150, ...],
#     'curb-weight': [3504, 3693, 3436, ...],
#     'engine-size': [130, 165, 150, ...],
#     'highway-mpg': [18, 15, 18, ...]
# })
# y_data = pd.Series([35000, 42000, 39000, ...])

# 1. Define Feature Columns and Target Variable
feature_cols = ['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']
X = x_data[feature_cols]
y = y_data

# 2. Create a Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Step 1: Feature Scaling
    ('ridge', Ridge())             # Step 2: Ridge Regression
])

# 3. Define the Parameter Grid for GridSearchCV
# Using a combination of linear and logarithmic scales for alpha
linear_alphas = 10 * np.arange(1, 1001)  # 10, 20, ..., 10,000
log_alphas = np.logspace(3, 5, 100)      # 1000 to 100,000 in log scale
alpha_range = np.concatenate([linear_alphas, log_alphas])

parameters2 = {
    'ridge__alpha': alpha_range,          # Regularization strength
    'scaler': [StandardScaler(), 'passthrough']  # Whether to scale features
}

# 4. Initialize GridSearchCV
Grid2 = GridSearchCV(
    estimator=pipeline,
    param_grid=parameters2,
    cv=4,                                      # 4-fold cross-validation
    scoring='neg_mean_squared_error',          # Negative MSE as scoring metric
    verbose=1,                                 # Verbosity level for monitoring
    n_jobs=-1                                   # Utilize all CPU cores
)

# 5. Fit GridSearchCV on the Data
Grid2.fit(X, y)

# 6. Retrieve the Best Parameters
best_alpha = Grid2.best_params_['ridge__alpha']
best_scaler = Grid2.best_params_['scaler']

# 7. Output the Best Parameters
print(f"Best alpha: {best_alpha}")
print(f"Scaler used: {best_scaler}")

# 8. Initialize the Best Ridge Regression Model with Best Parameters
best_ridge_model = Pipeline([
    ('scaler', best_scaler),
    ('ridge', Ridge(alpha=best_alpha))
])

# 9. Fit the Best Model on the Entire Dataset
best_ridge_model.fit(X, y)

# 10. (Optional) Evaluate the Model's Performance on the Training Data
train_score = best_ridge_model.score(X, y)
print(f"Training R² Score of the Best Model: {train_score}")

# 11. (Optional) Evaluate the Model's Performance on Test Data (if available)
# Assuming you have separate test data: X_test and y_test
# test_score = best_ridge_model.score(X_test, y_test)
# print(f"Test R² Score of the Best Model: {test_score}")
````
